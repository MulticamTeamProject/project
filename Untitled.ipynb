```python
data = os.listdir()
data
```

```python
name = ['Manila', 'Bohol', 'Cebu']
col = ['City', 'Name', 'Score', 'Description']
for i, x in enumerate(data):
    df = pd.read_csv(x).copy()
    df['City'] = name[i]
    df.reindex(columns=col)
    print(df)
    result = []
    for y in df.index:
        temp = []
        temp.append(df.iloc[y,0])
        temp.append(df.iloc[y,1])
        temp.append(df.iloc[y,2])
        temp.append(df.iloc[y,3])
        result.append(temp)
    df1 = pd.DataFrame(result, columns=col)
    df1.to_csv('dummy_'+data[i], encoding='utf-8')
    print(df1)
```

```python
df1 = pd.read_csv('dummy_마닐라.csv')
df2 = pd.read_csv('dummy_보홀.csv')
df3 = pd.read_csv('dummy_세부.csv')
df1['City'] = 'Kyoto'
df2['City'] = 'Tokyo'
df3['City'] = 'Osaka'
```

```python
df = pd.concat([df1,df2,df3])
df.index = np.arange(len(df))
df.to_csv('종합.csv', encoding='utf-8')
```
